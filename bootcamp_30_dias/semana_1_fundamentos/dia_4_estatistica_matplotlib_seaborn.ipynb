## üìã Parte 1 ‚Äî Estat√≠stica
### Para a lista dados = [12, 15, 20, 22, 18, 17, 25, 30], calcule:

- M√©dia 

- Mediana

- Desvio padr√£o

- Vari√¢ncia

- Correla√ß√£o

Em seguida, crie duas listas:

```python
import numpy as np
```

```python
# Lista de dados
dados = [12, 15, 20, 22, 18, 17, 25, 30]
dados_np = np.array(dados)
```

```python
# M√©dia
media = np.mean(dados_np)
print("M√©dia:", media)

# Mediana
mediana = np.median(dados_np)
print("Mediana:", mediana)

# Desvio padr√£o
desvio_padrao = np.std(dados_np)
print("Desvio padr√£o:", desvio_padrao)

# Vari√¢ncia
variancia = np.var(dados_np)
print("Vari√¢ncia:", variancia)

# Correla√ß√£o da lista consigo mesma (auto-correla√ß√£o)
correlacao_auto = np.corrcoef(dados_np)
print("Correla√ß√£o (auto):")
print(correlacao_auto)
```

# Correla√ß√£o entre duas listas
-  Plote um gr√°fico de linha para os valores de y em fun√ß√£o de x.

- Plote um histograma da lista dados.


```python
# Listas x e y
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 5, 4, 5])

# Correla√ß√£o entre x e y
correlacao_xy = np.corrcoef(x, y)
print("Correla√ß√£o entre x e y:")
print(correlacao_xy)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Estilo visual do Seaborn
sns.set(style="whitegrid")
# Dados
dados = [12, 15, 20, 22, 18, 17, 25, 30]
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 5, 4, 5])

# Estilo do Seaborn
sns.set(style="whitegrid")

# Gr√°fico de linha com Seaborn
plt.figure(figsize=(6, 4))
sns.lineplot(x=x, y=y, marker='o', color='blue')
plt.title("Gr√°fico de Linha: y em fun√ß√£o de x")
plt.xlabel("x")
plt.ylabel("y")
plt.tight_layout()
plt.savefig("grafico_linha.png")  # Salva como imagem
plt.show()

# Histograma com Seaborn
plt.figure(figsize=(6, 4))
sns.histplot(dados, bins=5, kde=True, color='green')# kde = true add curva de densidade
plt.title("Histograma dos Dados")
plt.xlabel("Valor")
plt.ylabel("Frequ√™ncia")
plt.tight_layout() # evita que o texto fique cortados
plt.savefig("histograma_dados.png")  # Salva como imagem
plt.show()
```
